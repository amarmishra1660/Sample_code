```python
import re
article_text_new = []
```

```python
for i in range(0,len(article_text)):
    article_text_new.append(re.sub('\n', ' ',article_text[i]))
```

```python
formatted_article_text = []
for i in range(0,(len(article_text_new)-1)):
    formatted_article_text.append(re.sub('[^a-zA-Z]', ' ', article_text_new[i] ))
    formatted_article_text[i] = re.sub(r'\s+', ' ', formatted_article_text[i])
```

```python
len(article_text_new)
```

```python
import numpy as np
import pandas as pd
import nltk
nltk.download('punkt') # one time execution
```

```python
from nltk.tokenize import sent_tokenize
sentences = []
for s in article_text_new:
  sentences.append(sent_tokenize(s))

sentences = [y for x in sentences for y in x] # flatten list
```

```python
sentences
```

```python
# remove punctuations, numbers and special characters
clean_sentences = pd.Series(sentences).str.replace("[^a-zA-Z]", " ")

# make alphabets lowercase
clean_sentences = [s.lower() for s in clean_sentences]
```

```python
clean_sentences
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
# function to remove stopwords
def remove_stopwords(sen):
    sen_new = " ".join([i for i in sen if i not in stop_words])
    return sen_new
```

```python
clean_sentences = [remove_stopwords(r.split()) for r in clean_sentences]
```

```python
# Extract word vectors
word_embeddings = {}
f = open('glove.6B.100d.txt', encoding='utf-8')
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    word_embeddings[word] = coefs
f.close()
```

```python
len(word_embeddings.get('got'))
```

```python
sentence_vectors = []
for i in clean_sentences:
  if len(i) != 0:
    v = sum([word_embeddings.get(w, np.zeros((100,))) for w in i.split()])/(len(i.split())+0.001)
  else:
    v = np.zeros((100,))
  sentence_vectors.append(v)
```

```python
# similarity matrix
sim_mat = np.zeros([len(sentences), len(sentences)])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_mat[3][1]
```

```python
for i in range(len(sentences)):
  for j in range(len(sentences)):
    if i != j:
      sim_mat[i][j] = cosine_similarity(sentence_vectors[i].reshape(1,100), sentence_vectors[j].reshape(1,100))[0,0]
```

```python
import networkx as nx

nx_graph = nx.from_numpy_array(sim_mat)
scores = nx.pagerank(nx_graph)
```

```python
ranked_sentences = sorted(((scores[i],s) for i,s in enumerate(sentences)), reverse=True)
```

```python
# Extract top 10 sentences as the summary
for i in range(10):
  print(ranked_sentences[i][1])
```
